## чтение

```python
import csv

data = []

with open(PATH, 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file, delimiter=';')
    for row in reader:
        data.append(row)
```

```python
data[:4]
```

## 1 задание
Вывести в понятном виде иерархию команд, т.е. департамент и все команды, которые входят в него

```python
departments = set([el['Департамент'] for el in data])
hierarchy = {dept: set() for dept in departments}
```

```python
departments
```

```python
hierarchy
```

```python
for el in data:
    branch = el['Отдел']
    department = el['Департамент']
    if branch not in department:
        hierarchy[department].add(branch)
```

```python
for department, branches in hierarchy.items():
    print(department + ':')
    for branch in branches:
        print(f'    {branch}')
    print()
```

## 2 задание
Вывести сводный отчёт по департаментам: название, численность, "вилка" зарплат в виде мин – макс, среднюю зарплату

```python
report = {}
```

```python
data[0]
```

```python
for el in data:

    department = el['Департамент']
    salary = float(el['Оклад'])

    if department not in report:
        report[department] = {
            'count': 0,
            'min': float('inf'),
            'max': -float('inf'),
            'total': 0,
        }

    report[department]['count'] += 1
    if salary < report[department]['min']:
        report[department]['min'] = salary
    if salary > report[department]['max']:
        report[department]['max'] = salary
    report[department]['total'] += salary
```

```python
report
```

```python
for dept, _ in report.items():
    report[dept]['average'] = round(report[dept]['total'] / report[dept]['count'], 0)
    del report[dept]['total']
```

```python
report
```

```python
for department, info in report.items():
    print(f"{department} | {info['count']} | {info['min']} | {info['max']} | {info['average']}")
```

## 3 задание
Сохранить сводный отчёт из предыдущего пункта в виде csv-файла. При этом необязательно вызывать сначала команду из п.2


```python
for k, v in report.items():
    v.update({'Department': k})

final_report = list(report.values())
```

```python
final_report
```

```python
list(report.keys())
```

```python
import csv
header = ['Department', 'count', 'min', 'max', 'average']

name_file = 'report.csv'
with open(f'{name_file}', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=header)
    writer.writeheader()
    writer.writerows(final_report)
```
